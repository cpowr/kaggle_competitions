```python
import pandas as pd
test_dataset = pd.read_csv('test.csv')
train_dataset = pd.read_csv('train.csv')
```

```python
test_dataset.head()
```

```python
from pycaret.utils import enable_colab
enable_colab()
```

```python
train_dataset.info()
```

```python
from pycaret.classification import *

#pyc_classif = setup(train_dataset, silent=True, target='label', use_gpu = True, session_id = 1)
pyc_classif = setup(train_dataset, use_gpu=True, silent=True, target='label', session_id = 7)
```

```python
models()
```

```python
lightgbm_model = create_model('lightgbm')
```

```python
et_model = create_model('et')
```

```python
knn_model = create_model('knn')
```

```python
blender_soft = blend_models(estimator_list = [knn_model,et_model,lightgbm_model], method = 'soft')
```

```python
blender_final = finalize_model(blender_soft)
```

```python
save_model(blender_soft, 'blended_20211029')
```

```python
save_model(blender_final, 'blended_final_20211029')
```

```python
predictions_test = predict_model(blender_final, data = test_dataset)
```

```python
predictions_test
```

```python
save_model(predictions_test, 'mnist_predictions_20211029')
```
